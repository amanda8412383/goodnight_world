```python
import pandas as pd
import numpy as np
import csv
```

## original csv

```python
ceo_name = pd.read_csv('ceo name.csv', engine = "python", index_col=False, skiprows = 1, 
    encoding ='utf-8', header=None, na_values = '-')
ceo_name.columns  = ceo_name.columns  = ['CO_PER_ROL', 'YEAR', 'CUSIP', 'CONAME', 'EXEC_FNAME', 'EXEC_FULLNAME', 'EXEC_LNAME', 'EXEC_MNAME',	'ceo_nameANN']
ceo_name = ceo_name.drop_duplicates()
ceo_name = ceo_name[ceo_name.YEAR == 2017]
ceo_name = ceo_name.filter(items=['CO_PER_ROL',  'YEAR', 'EXEC_FULLNAME'])

ceo_name.head()
```

## result of 2017 search

```python
ceo_2017 = pd.read_csv('CEO_2017_pre.csv', engine = "python", index_col=False, skiprows = 1, 
    encoding ="ISO-8859-1", header=None, na_values = '-')
ceo_2017.columns  = ['CO_PER_ROL', 'YEAR', 'CUSIP', 'CONAME', 'EXEC_FNAME', 'EXEC_LNAME', 'TWITTER', 'T_NAME', 'Search Date', 'company related', 'personal related', 'Note', 'Note2']
ceo_2017 = ceo_2017.filter(items=['CO_PER_ROL', 'YEAR', 'T_NAME'])
ceo_2017.loc[ceo_2017.T_NAME.isna()]

ceo_2017.head()

```

## merge to solve full name missed

```python
ceo_tweet_2017 = pd.merge(ceo_name, ceo_2017,on=['CO_PER_ROL',  'YEAR'], how='outer')
ceo_tweet_2017.loc[ceo_tweet_2017.T_NAME.isna()]
ceo_tweet_2017['CO_PER_ROL'].nunique() == ceo_tweet_2017['CO_PER_ROL'].count()
ceo_tweet_2017.head()
```

```python
#CHECK MISSING VALUE
see = ceo_tweet_2017[ceo_tweet_2017.T_NAME.isna()]
see

#COUNT RATIO 
ceo_tweet_2017[['T_NAME']] = ceo_tweet_2017[['T_NAME']].fillna(value='X')
ceo_tweet_2017 = ceo_tweet_2017.drop_duplicates()


#count = ceo_tweet_2017[ceo_tweet_2017.T_NAME == 'X'].value_counts()


# find = ceo_tweet_2017.loc[ceo_tweet_2017['T_NAME'] != 'X']['T_NAME'].count()
find2= len (ceo_tweet_2017.loc[ceo_tweet_2017['T_NAME'] != 'X'].index)

# miss = ceo_tweet_2017.loc[ceo_tweet_2017['T_NAME'] == '7']['T_NAME'].count()
miss2= len (ceo_tweet_2017.loc[ceo_tweet_2017['T_NAME'] == 'X'].index)

ratio = find2/miss2 * 100 

#eo_tweet_2017.loc[ceo_tweet_2017['T_NAME'] == '7']
ratio
```
